#### 문제 1. 본부장에게 보고할 월평균 광고비(백만원)의 95%의 신뢰구간을 구하라 구간의 하한값과 상한값은 각각 얼마인가
- 하한값 : 107.6만원
- 상한값 : 187.07만원

```python
my_plot.kdeplot(origin, '광고비(백만원)', meanline=True, clevel=0.95)
```

#### 문제 2. 같은 방식으로 월평균 매출액(억원)의 95% 신뢰구간을 구하라. 하한값과 상한값은?
- 하한값 : 265.74억원
- 상한값 : 810.60억원

```python
my_plot.kdeplot(origin, '매출액(억원)', meanline=True, clevel=0.95)
```

#### 문제 3. 광고비와 매출액 분포를 하나의 그래프로 시각화하라. 그래프를 통해 표시되는 표본평균은?
- 광고비 : 147.33만원
- 매출액 : 538.17억원

```python
df1= origin.melt(id_vars='월', value_vars=['광고비(백만원)','매출액(억원)'], var_name='구분', value_name='값')
df1.head()
```

```python
my_plot.kdeplot(df1, '값',hue='구분', meanline=True, clevel=0.95)
```
