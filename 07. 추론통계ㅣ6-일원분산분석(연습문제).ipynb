#### 문제 1 - 세가지 읽는 방법에 통계적으로 유의한 차이가 있는지 검정하라

```python
df1 = load_data('reading')
df1
```

```python
my_stats.anova_oneway(df1, y='점수', between='독서방법')
```

- 검정 선택 : test=welch_anova → 등분산성을 위반하여 Welch ANOVA 검정으로 진행
- 유의성 : F(2, 48.204)=10.787, p=0.000<0.05 → 귀무가설 기각, 세 독서방법 중 적어도 한 방법의 평균 점수는 다르다.
- 효과 크기 : np2=0.283 → Large, 독서방법 차이가 점수 변동의 약 28%를 설명, 통계적 유의성을 넘어 실질적으로도 큰 효과를 보였다.
- 다음 단계 : ANOVA는 '어딘가 차이가 있다'까지만 말해주므로, 어느 독서방법 쌍이 독서평가 점수가 다른지는 사후검정 등으로 확인 필요

#### 문제 2- 상권 종류별 2023년 한해 동안의 평균 매출액은 어떤 차이를 보이는지 분석하라

```python
df2 = load_data('commercial')
df2.head()
```

```python
my_stats.anova_oneway(df2, y='sales_amount', between='type')
```

- 검정 선택 : test=welch_anova → 등분산성을 위반하여 Welch ANOVA 검정으로 진행
- 유의성 : F(3, 110.239)=161.433, p=0.000<0.05 → 귀무가설 기각, 네 상권 중 적어도 한 상권의 평균 매출액은 다르다.
- 효과 크기 : np2=0.228 → Large, 상권종류별 2023년 한해 동안의 평균매출액 차이가 약 22.8%를 설명, 통계적 유의성을 넘어 실질적으로도 큰 차이를 보였다.
- 다음 단계 : ANOVA는 '어딘가 차이가 있다'까지만 말해주므로, 어느 한 상권의 매출평균이 다른지는 사후검정 등으로 확인 필요
