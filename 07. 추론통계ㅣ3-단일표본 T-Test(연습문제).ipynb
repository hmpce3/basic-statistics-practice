# [LAB 07] 3. 단일표본 T-Test
## #연습문제 - 빵집의 양심

```python
from hossam import load_data
from helpers import my_stats
```

```python
df=load_data('bread')
df.head()
```

- 빵의 평균무게가 500g과 차이가 있는지 확인
- 귀무가설 : 빵의 평균무게가 500g과 차이가 없다.
- 대립가설 : 빵의 평균무게가 500g과 차이가 있다.

```python
my_stats.test_1sample(df, column='무게', popmean=500)
```

- One-sample t-test 양측검정 결과 t값은 -37.574, p-value는 0으로 나타났다. p<0.05이므로 귀무가설을 기각하고 대립가설을 채택한다. 이는 빵의 평균무게가 500g과 차이가 있다는 것을 의미한다.
- 좌측 단측검정 결과 p-value는 0으로 나타났다. p<0.05이므로 귀무가설을 기각하고 대립가설을 채택한다. 즉, 빵의 평균무게가 500g보다 작다.
- 우측 단측검정 결과 p-value는 1로 나타났다. p>=1이므로 귀무가설을 채택한다. 즉, 빵의 평균무게는 500g보다 크지 않다.
- 빵의 평균무게가 500g보다 작다는 것을 확인할 수 있다.
